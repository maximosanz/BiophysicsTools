```python
from CircularDichroism import CD_Unfolding
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Input file
Fn = "Protein_CD_Melting.csv"

X = np.array([ l.split(',') for l in open(Fn) ],dtype=np.float64).T

T = X[0]
Y = X[1]
```

```python
CD = CD_Unfolding(T,Y,T_Units="C",Energy_Units="kcal")

El = CD.Gibbs_Helmholtz_Eq(CD.T,*CD.Pars)
```

```python
plt.plot(T,El,color='k',lw=1)
plt.scatter(T,Y,s=4,color='k')
plt.suptitle("Before Fitting")
plt.xlabel("Temperature (C)")
plt.ylabel("Molar ellipticity at 222 nm")
plt.show()
```

```python
CD.Fit()
El = CD.Gibbs_Helmholtz_Eq(CD.T,*CD.Pars)
```

```python
plt.plot(T,El,color='k',lw=1)
plt.scatter(T,Y,s=4,color='k')
plt.suptitle("After Fitting")
plt.xlabel("Temperature (C)")
plt.ylabel("Molar ellipticity at 222 nm")
plt.show()
```
